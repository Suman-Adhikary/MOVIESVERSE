```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
data = pd.read_csv('/content/Hollywood_with_Poster.csv')
data = data.drop_duplicates()
data.head(10)
```

```python
data[data['original_title']=='Jurassic World']['Poster'].values[0]
```

```python
data = data.reset_index()
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
characters_to_remove = ',.|:?'
data['Combined'] = data['Combined'].str.replace('[' + characters_to_remove + ']', ' ', regex=True)
```

```python
data['Combined'].head(10)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
data['Combined'] = data['Combined'].fillna('')
tfidf_matrix = tfidf.fit_transform(data['Combined'])
```

```python
print(tfidf_matrix)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
data.tail()
```

```python
indices = pd.Series(data.index, index=data['original_title']).drop_duplicates()
```

```python
indices
```

```python
indices.to_pickle("Hollywood.pkl")
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    return data[['id', 'original_title', 'Poster']].iloc[movie_indices]
```

```python
get_recommendations('Jurassic World')
```
